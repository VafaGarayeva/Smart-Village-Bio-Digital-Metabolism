```python
import pandas as pd
import numpy as np

# Sətir sayı (Analiz üçün 1500 sətir kifayətdir)
n = 1500

# Regionlar və Sensor növləri
regions = ['Şimal-Meşə', 'Cənub-Otlaq', 'Şərq-Məhsuldar', 'Qərb-Yaşayış']

# Tarix aralığı (Son 6 ayın saatlıq datası)
dates = pd.date_range(start='2025-09-01', periods=n, freq='4h')

data = {
    'Timestamp': dates,
    'Region': np.random.choice(regions, n),
    
    # 1. BİOLOGİYA: Torpaq Sağlamlığı (Nitrogen/Azot səviyyəsi 10-60 mg/kg)
    'Soil_Nitrogen_mg_kg': np.random.uniform(10, 60, n),
    
    # 2. ENERJİ: Tullantıdan alınan Biogaz (kW)
    'Waste_to_Biogas_kW': np.random.uniform(50, 450, n),
    
    # 3. TƏHLÜKƏSİZLİK: Akustik Anomaliya (0=Sakitlik, 1=Təhlükə/Heyvan stresi)
    'Acoustic_Anomaly_Score': np.random.beta(a=2, b=5, size=n), 
    
    # 4. İQTİSADİYYAT: Ağıllı sistemin qənaət etdiyi məbləğ (AZN)
    'Efficiency_Savings_AZN': np.random.uniform(20, 150, n),
    
    # 5. SU: Sızma dərəcəsi (Litr/saat)
    'Water_Leakage_Liters': np.random.exponential(scale=5, size=n)
}

df = pd.DataFrame(data)

# Məntiqli asılılıqlar yaradaq (Məsələn: Azot azdırsa, Efficiency düşür)
df.loc[df['Soil_Nitrogen_mg_kg'] < 20, 'Efficiency_Savings_AZN'] *= 0.7

# CSV olaraq yadda saxla
df.to_csv('Smart_Village_Master_Data.csv', index=False)
print("Fayl hazırdır: Smart_Village_Master_Data.csv")
```
